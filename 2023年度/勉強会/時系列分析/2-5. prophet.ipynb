```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/udemy_time_series/energydata_complete.csv")
df
```

```python
df["date"] = pd.to_datetime(df["date"])
df
```

```python
df.dtypes
```

```python
df = df.set_index("date").resample("H").mean()
df
```

```python
df = df.rename(columns={"Appliances":"y"})
df.head()
```

```python
y_train = df[["y"]].loc[:"2016-04-30", :].reset_index()
y_test = df[["y"]].loc["2016-05-01":, :].reset_index()
```

```python
y_train
```

```python
y_test
```

```python
y_train = y_train.rename(columns={"date":"ds"})
y_test = y_test.rename(columns={"date":"ds"})
```

```python
y_train
```

```python
y_test
```

```python
from prophet import Prophet
```

```python
m = Prophet()
m.fit(y_train)
```

```python
future = m.make_future_dataframe(periods = len(y_test), freq = 'H')
future
```

```python
pred = m.predict(future)
pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
result_df = pd.merge(pred[["ds", "yhat"]].astype(str), y_test.astype(str), on="ds", how="inner")
result_df
```

```python
rmse = np.sqrt(mean_squared_error(result_df["y"], result_df["yhat"]))
print("RMSE:{:.2f}".format(rmse))
```

```python
result_df["ds"] = pd.to_datetime(result_df["ds"])
result_df["y"] = result_df["y"].astype(float)
result_df["yhat"] = result_df["yhat"].astype(float)
result_df.set_index("ds").plot()
```
